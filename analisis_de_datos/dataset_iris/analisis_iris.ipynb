````python
import csv

# Inicializamos la estructura del dataset con listas vacías
dataset = [ [], [], [], [], [] ]
# Definimos los nombres de las mediciones
measurements = [ "largo del sépalo", "ancho del sépalo", "largo del pétalo", "ancho del pétalo" ]

# Abrimos el archivo CSV y leemos los datos
with open("sample_data/iris.csv") as f:
    for row in csv.reader(f):
        if len(row) == 5:  # Solo procesamos filas con 5 columnas, como se espera
            for i in range(4):  # Las 4 mediciones deben convertirse a float
                dataset[i].append(float(row[i]))
            # La posición 4 es el tipo de iris, que se mantiene como cadena de texto
            dataset[4].append(row[4])

# Mostramos las primeras 5 filas para verificar la carga de datos
print("Primeras 5 filas del dataset:")
for i in range(5):
    print([dataset[j][i] for j in range(5)])
````

Ahora que cada uno de los elementos de `dataset` contiene una de las columnas del conjunto de datos **Iris**, el segundo ejercicio consiste en calcular la **media** y la **desviación estándar** de los primeros 4 elementos (listas) de `dataset`.

Primero, necesitamos definir un par de funciones que calculen la **media** y la **desviación estándar** de una lista de valores.

Comencemos con la media.


````python
def mean(x):
    return sum(x) / len(x)
````

La función `sum()` es muy útil. Suma todos los valores de una lista dada. Luego, dividimos esta suma por la longitud de la lista `x`, calculando así el **valor medio** de manera efectiva.

Por otro lado, la **desviación estándar** se puede ver como la raíz cuadrada de la **media de las diferencias al cuadrado** entre cada elemento de `x` y la media.

Desglosemos esto con un pequeño **ejemplo**:


````python
x = [ 5, 3, 15, 4, 1 ]
u = mean(x)
print(u)
````

La diferencia al cuadrado de cada elemento respecto a la media `u`  
(que hemos calculado usando la función `mean()` definida anteriormente) se expresa como:


````python
diff = [ (x_i - u) ** 2 for x_i in x ]
print(diff)
````

Una vez que hemos calculado esta diferencia, podemos calcular su **media**.  
La función `mean()` que ya hemos creado nos será útil para ello.  
Finalmente, calculamos su **raíz cuadrada**.


````python
def std(x):
    u = mean(x)
    return (mean([ (x_i - u) ** 2 for x_i in x ])) ** 0.5

print(std(x))
````

Ahora, podemos aplicar estas dos funciones a nuestras columnas.


````python
for i, m in enumerate(measurements):
    print(f"{m} mean {mean(dataset[i]):.4f} std {std(dataset[i]):.4f}")
````

Repitiendo el análisis anterior, pero esta vez para cada tipo de **Iris** por separado.

Primero, identifiquemos los diferentes tipos de **Iris** disponibles.  
Podríamos sentirnos tentados a crear una lista de valores como la siguiente:


````python
tipos_iris = [ 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica' ]
print(tipos_iris)
````

Ahora podemos iterar sobre cada columna y, para cada columna, sobre cada tipo de **Iris**.  
Luego, seleccionamos todos los valores correspondientes a esa **medición** y a ese **tipo de Iris**  
y calculamos la **media** y la **desviación estándar**.


````python
for i, medida in enumerate(measurements):
    print(medida)
    for tipo_iris in tipos_iris:
        # Para cada medición y cada tipo de iris, construimos una lista de valores
        valores = [ v for v, t in zip(dataset[i], dataset[4]) if t == tipo_iris ]
        print(f"{tipo_iris} {mean(valores):.4f} {std(valores):.4f}")
    print()
````

Basándonos en los resultados anteriores, la medición que **más ayuda a diferenciar** entre los tres tipos de **Iris** es aquella que tiene las **medias más distantes** y las **desviaciones estándar más pequeñas**.

Esto corresponde a distribuciones de valores que están **bien separadas** (medias distantes)  
y cuyos valores se agrupan cerca de la media (desviación estándar pequeña).

A primera vista, la medición que más destaca es el **ancho del pétalo**. Sus medias están bien separadas (aproximadamente **0.2, 1.3 y 2.0**) y sus desviaciones estándar son bajas (todas por debajo de **0.27**).

Un apoyo visual puede ayudar a comprender mejor esta diferencia. Aunque no lo hemos introducido todavía, el siguiente código generará una **curva gaussiana o normal de campana**  y los histogramas reales para cada uno de los tres tipos de **Iris**, para cada medición.  Las medias y desviaciones estándar de las curvas son las que hemos calculado anteriormente.


Info adicional

**`norm(u, s).pdf(x)`**  
   - `scipy.stats.norm` crea una **distribución normal** con media `u` y desviación estándar `s`.  
   - `pdf(x)` calcula la **función de densidad de probabilidad (PDF)** para los valores en `x`, lo que genera la curva de campana.



````python
import matplotlib.pyplot as plt
#%matplotlib inline
from scipy.stats import norm
import numpy as np

# Definimos los colores para cada tipo de Iris
colores = ['b', 'g', 'r']

# Iteramos sobre cada una de las mediciones (longitud y ancho de sépalo y pétalo)
for i, medida in enumerate(measurements):
    plt.figure()  # Creamos una nueva figura para cada medición
    plt.title(f"Distribución de {medida}")  # Título del gráfico

    for tipo_iris, color in zip(tipos_iris, colores):
        # Extraemos los valores de la medición actual (i) solo para el tipo de Iris correspondiente
        valores = [v for v, t in zip(dataset[i], dataset[4]) if t == tipo_iris]
        # zip(dataset[i], dataset[4]) empareja los valores de la medición con los tipos de Iris
        # Luego, filtramos solo aquellos valores que corresponden al tipo_iris actual

        # Graficamos el histograma
        plt.hist(valores, density=True, alpha=0.2, color=color)

        # Calculamos la media y la desviación estándar
        u = mean(valores)
        s = std(valores)

        # Generamos un rango de valores en torno a la media para dibujar la curva de distribución normal
        x = np.linspace(u - 5 * s, u + 5 * s, 100)

        # Graficamos la distribución normal teórica basada en la media y desviación estándar calculadas
        plt.plot(x, norm(u, s).pdf(x), label=f"{tipo_iris}", color=color)

        # Etiquetas de los ejes
        plt.xlabel(f"{medida} (cm)")
        plt.ylabel("Densidad")

    plt.legend()
    plt.show()
````

Como era de esperar, las **curvas de campana** para el **ancho del pétalo** están bastante bien separadas. Lo mismo ocurre con la **longitud del pétalo**, que también es un buen separador entre las tres clases.  

En cambio, la **longitud y el ancho del sépalo** son **peores discriminadores**, es decir, **no son tan útiles** para diferenciar entre las tres clases de Iris.  

Finalmente, observa cuán diferente es **Iris Setosa** en comparación con **Iris Virginica** e **Iris Versicolor**. Esto hace que identificar si una flor es un **Iris Setosa** sea una tarea mucho más sencilla, en comparación con diferenciar entre **Versicolor** y **Virginica**.

# Análisis con pandas

````python
import pandas as pd 

archivo_malo = 'iris.data'
archivo_bueno = 'iris.csv'

df0 = pd.read_csv(archivo_malo)
# Se quitan TODOS los valores na's o vacios 
df0.dropna(how='all', inplace=True)
# Guardamos el archivo en otro archivo csv con nombre "iris.csv"
df0.to_csv(archivo_bueno, index=False)
# Guardamos en un dataframe "df" el archivo bueno y le ponemos nombres a las columnas
df = pd.read_csv(archivo_bueno, names = ['Longitud sépalo', 'Ancho sépalo',
'Longitud pétalo', 'Ancho pétalo', 'Especie'])
print(df)
````

````python
import csv

with open("iris.csv") as f:
    for cols in csv.reader(f):
        print(cols[0], cols[2])
````

Con los datos en el dataframe "df" definimos las funciones "media" y "desvest" para obtener la media y la desviacion estandar respectivamente.

````python
def media(df, columnas):
    medias = {}
    for col in columnas:
        suma = 0
        numero = 0
        for valor in df[col]:
            suma += valor
            numero += 1
        medias[col] = suma / numero 
    return medias

def desvest(df, columnas):
    desviaciones = {}
    medias = media(df, columnas)
    for col in columnas:
        suma_cuadrados = 0
        numero = 0
        for valor in df[col]:
            suma_cuadrados += (valor - medias[col]) ** 2
            numero += 1
        desviaciones[col] = (suma_cuadrados / (numero - 1)) ** 0.5 
    return desviaciones
````

````python
print("La media de todos los datos son:", media(df, ['Longitud sépalo', 'Ancho sépalo',
'Longitud pétalo', 'Ancho pétalo']))
print("Mientras que la desviación estándar es:", desvest(df, ['Longitud sépalo', 'Ancho sépalo',
'Longitud pétalo', 'Ancho pétalo']))
````

````python
datos_setosas = df[df['Especie'] == 'Iris-setosa']
print(datos_setosas)
````

Se eligieron únicamente las columnas con valores numéricos para ocupar la función "mean" y "std" para obtener el promedio y desviación estándar respectivamente. Después se imprimió el resultado. 

````python
print("La media los datos para la iris-setosa son:", media(datos_setosas, ['Longitud sépalo', 'Ancho sépalo',
'Longitud pétalo', 'Ancho pétalo']))
print("Mientras que la desviación estándar para la iris-setosa es:", desvest(datos_setosas, ['Longitud sépalo', 'Ancho sépalo',
'Longitud pétalo', 'Ancho pétalo']))
````

# Analisis con PCA en Scikit-learn

````python
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
````

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
````

Vemos que la variable X es un array con 4 componentes por elemento. 

````python
X
````

Entonces ocupando el ejemplo de la paquetería scikitlearn 
```python
from sklearn.preprocessing import StandardScaler

  ```
Definimos la variable Z estandarizada.

````python
scaler = StandardScaler()
Z = scaler.fit_transform(X)
````

Aplicamos PCA con todas las componentes

````python
pca_full = PCA(n_components=4)
X_pca_full = pca_full.fit_transform(Z)
````

Obtenemos la varianza explicada acumulada

````python
explained_variance = pca_full.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
````

Visualizamos la valrianza aculumada.

````python
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance)
plt.xlabel("Número de componentes")
plt.ylabel("Varianza explicada acumulada")
plt.grid()
plt.show()
````

Graficamos la varianza acumulada y la valianza por componente de forma más visible.

````python
plt.figure(figsize=(8, 6))
plt.bar(range(1, 5), explained_variance * 100, alpha=0.7, label="Varianza por componente")
plt.plot(range(1, 5), cumulative_variance * 100, marker="o", linestyle="--", color="r", label="Varianza acumulada")
plt.xticks(range(1, 5))
plt.xlabel("Número de Componentes")
plt.ylabel("Varianza Explicada")
plt.legend()
plt.grid()
plt.show()
````

De manera que vemos que con dos variables podemos describir muy bien de qué flor estamos tratando.

Entonces aplicamos ahora si PCA tal y como lo vimos en la clase para reducir la dimensionalidad a dos variables. 

````python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(Z)
df_pca1 = pd.DataFrame(X_pca,
columns=['PC{}'.format(i+1)
for i in range(2)])
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris['target'])
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
````

Imprimimos los valores de varianza explicada.

````python
explained_variance = pca.explained_variance_ratio_
print(f'Varianza explicada por cada componente: {explained_variance}')
print(f'Varianza total explicada: {np.sum(explained_variance)}')
````

Hacemos más bonita la gráfica.

````python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=iris.target_names[y], palette="viridis")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend(title="Especies")
plt.grid()
plt.show()
````
